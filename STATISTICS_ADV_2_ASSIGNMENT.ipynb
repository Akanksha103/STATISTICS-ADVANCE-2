Q1. What is hypothesis testing in statistics?

Ans. In statistics, hypothesis testing is a formal process used to make decisions or draw conclusions about a population based on sample data.

It’s like a court trial for your data — you start with an assumption (the null hypothesis) and then check whether the evidence (your sample data) is strong enough to reject it in favor of an alternative hypothesis.



Q2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?

Ans. 1. Null Hypothesis (H₀)

Definition: The default assumption that there is no effect, no difference, or no relationship in the population.

Purpose: Acts as the starting point; we assume it’s true until evidence suggests otherwise.

Example: A medicine has no effect on blood pressure. (Mean change = 0)

**The null hypothesis (H₀) is the default claim that there is no significant difference, effect, or relationship in the population. It assumes that any variation we observe in the sample is purely due to chance. In hypothesis testing, we start by assuming H₀ is true and only reject it if the data provides strong evidence against it. For example, saying “The average exam score of this class is exactly 75” is a null hypothesis.

**The alternative hypothesis (H₁ or Ha), on the other hand, is the statement that challenges the null hypothesis. It claims that there is a real difference, effect, or relationship present in the population. It is what the researcher often hopes to demonstrate. For example, saying “The average exam score of this class is not 75” would be an alternative hypothesis.

In short, the null hypothesis is the “no change” or “no effect” position, while the alternative hypothesis is the “there is change” or “there is an effect” position. We test the sample evidence to decide whether to reject the null hypothesis in favor of the alternative.



Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test?


Ans. The significance level (denoted by α) in hypothesis testing is the threshold you set to decide how much risk you are willing to take of making a Type I error — rejecting the null hypothesis when it is actually true.

Meaning

It represents the maximum probability of wrongly rejecting H₀.
Common choices are α = 0.05 (5%) or α = 0.01 (1%).
For example, if α = 0.05, you’re accepting a 5% chance of making a wrong rejection.

 **Role in Deciding the Outcome

Compare p-value to α:

If p-value ≤ α → Reject H₀ (evidence is strong enough against the null).

If p-value > α → Fail to reject H₀ (evidence is not strong enough).

Sets the cut-off for critical values:

In z-tests, t-tests, etc., α determines the critical region (the extreme values of the test statistic where we reject H₀).

Q4. What are Type I and Type II errors? Give examples of each?

Ans. 1. Type I Error (False Positive)

Definition: Rejecting the null hypothesis (H₀) when it is actually true.

Meaning: You think there’s an effect when there isn’t.

Probability: Equal to the significance level α.

Example:

A medical test says a healthy person has a disease.

H₀: The person is healthy.

H₁: The person has the disease.

Type I error → Declaring them sick when they’re actually healthy.

2. Type II Error (False Negative)

Definition: Failing to reject the null hypothesis when it is actually false.

Meaning: You miss detecting a real effect.

Probability: Denoted by β (Power of a test = 1 − β).

Example:

A medical test fails to detect a disease in a sick person.

H₀: The person is healthy.

H₁: The person has the disease.

Type II error → Saying they’re healthy when they’re actually sick.

Q5. What is the difference between a Z-test and a T-test? Explain when to use
each?

Ans. 1. Z-test

When to use:

Large sample size (n ≥ 30).

Population standard deviation (σ) is known.

Data is normally distributed or sample is large enough for Central Limit Theorem to apply.

Example: Testing whether the average weight of 500 apples from a farm is equal to the claimed average of 150 g, when you know the population standard deviation.

2. T-test

When to use:

Small sample size (n < 30).

Population standard deviation is unknown — we use sample standard deviation instead.

Data is approximately normally distributed.

Example: Testing whether the average score of a class of 15 students differs from the national average when the population standard deviation is not known.

**Key difference :

Z-test → Big sample + σ known.

T-test → Small sample + σ unknown.

Q6. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10      # number of trials
p = 0.5     # probability of success
size = 1000 # number of random samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(n+2)-0.5, edgecolor='black', alpha=0.7)
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q7. : Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population standard deviation
sigma = 0.5

# Hypothesized population mean
mu_0 = 50

# Calculate sample mean and size
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Calculate Z-statistic
z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Evidence suggests the mean is different from 50.")
else:
    print("Fail to reject the null hypothesis: Not enough evidence to say the mean is different from 50.")
```

Q8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate normal distribution data
np.random.seed(42)  # for reproducibility
mean_true = 50
std_dev = 5
n = 100
data = np.random.normal(mean_true, std_dev, n)

# Step 2: Calculate 95% confidence interval
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
alpha = 0.05
df = n - 1  # degrees of freedom
t_critical = stats.t.ppf(1 - alpha/2, df)

margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 3: Plot data and confidence interval
plt.figure(figsize=(8, 5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"Lower CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='--', label=f"Upper CI = {ci_upper:.2f}")

plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q9. : Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Step 1: Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Step 2: Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Step 3: Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores (Standardized Data)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

    return z_scores

# Example dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Run function
z_scores = calculate_and_plot_zscores(np.array(data))

print("First 5 Z-scores:", z_scores[:5])
```
